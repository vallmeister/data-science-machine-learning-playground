## Exercise 9
In this exercise, we are working with the Auto dataset. First, we remove missing values from the data set.

```python
auto = pd.read_csv('Auto.csv')
auto.describe(include='all')
```

```python
# b)and c)
print("{:15}|{:>15}|{:>15}|{:>20}".format("Predictor", "Range", "Mean", "Standard Deviation"))
for i in auto._get_numeric_data():
    print("{:15}|{:15}|{:15.2f}|{:20.2f}".format(i, auto[i].max() - auto[i].min(), auto[i].mean(), auto[i].std()))
```

```python
# d)
auto_modified = auto.drop(auto.index[[i for i in range(9, 85)]])
print("{:15}|{:>15}|{:>15}|{:>20}".format("Predictor", "Range", "Mean", "Standard Deviation"))
for i in auto_modified._get_numeric_data():
    print("{:15}|{:15}|{:15.2f}|{:20.2f}".format(i, auto_modified[i].max() - auto_modified[i].min(), auto_modified[i].mean(), auto_modified[i].std()))
```
